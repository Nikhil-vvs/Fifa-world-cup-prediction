```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_rows = 1000
pd.options.display.max_columns = 100
```

# Importing dataset

```python
direc="/home/angps/Documents/WorldCupPrediction/dataset/"
rankings = pd.read_csv(direc + 'fifa_ranking.csv')
matches = pd.read_csv(direc + 'results.csv')
squads = pd.read_csv(direc + '2018 FIFA World Cup Squads.csv')
```

```python
rankings = rankings[rankings["rank_date"]>="2011-08-24"]
rankings = rankings.replace({"IR Iran":"Iran","Korea Republic":"South Korea"})
rankings = rankings.loc[rankings["country_full"].isin(list(squads["Team"]))]
rankings=rankings.drop(["country_abrv","confederation"],axis=1)
rankings["rank_date"] = rankings["rank_date"].apply(lambda x: x[:-3])
rankings.reset_index(inplace=True,drop=True)
rankings.head()
```

```python
matches=matches[matches["date"]>="2011-08-24"]  #rankings available 1993/08/08 onwards
matches.reset_index(inplace=True,drop=True)
matches =  matches.replace({'Germany DR': 'Germany',"Korea Republic":"South Korea"})
matches=matches[matches["home_team"].isin(list(rankings["country_full"])) & matches["away_team"].isin(list(rankings["country_full"]))]
#matches["date"] = pd.to_datetime(matches["date"])
matches["date"] = matches["date"].apply(lambda x:x[:-3])
matches["impt"] = np.where(matches["tournament"]!="Friendly",1,0)  #1 if impt (non-friendly)
matches["neutral"] = np.where(matches["neutral"]==False,0,1)   #1 if neutral stadium
matches["country"]=np.where(matches["neutral"]==1,0,np.where(matches["country"]==matches["home_team"],2,1))  #0 -neutral, 1-away team, 2- home team
matches = matches.drop(["city","tournament"],axis=1)
matches.reset_index(inplace=True,drop=True)
matches.head()
```

```python
for i in matches.index:
    for j in rankings.index:
        if rankings.loc[j,"country_full"]==matches.loc[i,"home_team"] and rankings.loc[j,"rank_date"]==matches.loc[i,"date"]:
            matches.loc[i,"home_ranking"] = rankings.loc[j,"rank"]
            matches.loc[i,"home_total_points"] = rankings.loc[j,"total_points"]
            matches.loc[i,"home_prev_points"] = rankings.loc[j,"previous_points"]
            matches.loc[i,"home_rank_change"] = rankings.loc[j,"rank_change"]
            matches.loc[i,"home_cur_year_avg_weighted"] = rankings.loc[j,"cur_year_avg_weighted"]
            matches.loc[i,"home_last_year_avg_weighted"] = rankings.loc[j,"last_year_avg_weighted"]
            matches.loc[i,"home_2_year_avg_weighted"] = rankings.loc[j,"two_year_ago_weighted"]
            matches.loc[i,"home_3_year_avg_weighted"] = rankings.loc[j,"three_year_ago_weighted"]
            break
for i in matches.index:
    for j in rankings.index:
        if rankings.loc[j,"country_full"]==matches.loc[i,"away_team"] and rankings.loc[j,"rank_date"]==matches.loc[i,"date"]:
            matches.loc[i,"away_ranking"] = rankings.loc[j,"rank"]
            matches.loc[i,"away_total_points"] = rankings.loc[j,"total_points"]
            matches.loc[i,"away_prev_points"] = rankings.loc[j,"previous_points"]
            matches.loc[i,"away_rank_change"] = rankings.loc[j,"rank_change"]
            matches.loc[i,"away_cur_year_avg_weighted"] = rankings.loc[j,"cur_year_avg_weighted"]
            matches.loc[i,"away_last_year_avg_weighted"] = rankings.loc[j,"last_year_avg_weighted"]
            matches.loc[i,"away_2_year_avg_weighted"] = rankings.loc[j,"two_year_ago_weighted"]
            matches.loc[i,"away_3_year_avg_weighted"] = rankings.loc[j,"three_year_ago_weighted"]
            break
matches["diff_in_ranking"]=matches["home_ranking"]-matches["away_ranking"]
matches["home_mean_weighted_over_years"] = (matches["home_cur_year_avg_weighted"]+matches["home_last_year_avg_weighted"]+
                                            matches["home_2_year_avg_weighted"]+matches["home_3_year_avg_weighted"])/4
matches["away_mean_weighted_over_years"] = (matches["away_cur_year_avg_weighted"]+matches["away_last_year_avg_weighted"]+
                                            matches["away_2_year_avg_weighted"]+matches["away_3_year_avg_weighted"])/4
matches["diff_in_mean_weighted_over_years"] = matches["home_mean_weighted_over_years"]  - matches["away_mean_weighted_over_years"]
matches["Result"] = np.where(matches["home_score"]-matches["away_score"]==0,1,
                             np.where(matches["home_score"]-matches["away_score"]>0,2,0))  #0-away win,1-draw,2-home win

matches.head()
```
